```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['Home Broadband Adoption (Percentage of  Households)'], kde=True)
plt.title('Distribution of Home Broadband Adoption')
plt.xlabel('Home Broadband Adoption Percentage')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='No Home Broadband Adoption by Quartile', order=df['No Home Broadband Adoption by Quartile'].value_counts().index)
plt.title('Count of Zip Codes by No Home Broadband Adoption Quartile')
plt.xlabel('No Home Broadband Adoption Quartile')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
